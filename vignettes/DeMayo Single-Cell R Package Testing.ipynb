## Impute cell types using `adobo`

This is a manual marker-based cell type imputation method. Essentially, you tell the algorithm the genes that are markers of cell types in a particular cluster, and `adobo` tells you what is what.

See this link for more information: https://oscar-franzen.github.io/adobo/#cell-type-prediction

Also note that you need to set the `RETICULATE_PYTHON` environment variable; see the top of this file.

```r
# Make a SAMPLE single-cell dataset with Seurat.
# Ordinarily, you'd have this already.

random_st_obj <- function (n_genes, n_cells, sample_name_len = 25, max_counts = 100) {

    sample_gene_names <- sample(gene_names, n_genes, replace = FALSE)
    sample_cell_names_lambda <- (function (x) paste0(sample(c("A","C","T","G"), 25, replace = TRUE), collapse = ''))
    sample_cell_names <- sapply(1:n_cells, sample_cell_names_lambda)

    mtx <- matrix(sample(0:max_counts, n_genes * n_cells, replace = TRUE), nrow = n_genes)
    rownames(mtx) <- sample_gene_names
    colnames(mtx) <- sample_cell_names
    
    st_obj <- Seurat::CreateSeuratObject(mtx)

    # Replace all of these calls with 'st_obj <- Seurat::SCTransform(st_obj)'
    st_obj <- Seurat::NormalizeData(st_obj)
    st_obj <- Seurat::ScaleData(st_obj, verbose = FALSE)
    st_obj <- Seurat::FindVariableFeatures(st_obj, verbose = FALSE)

    # Visualize cells
    st_obj <- Seurat::RunPCA(st_obj, verbose = FALSE)
    st_obj <- Seurat::RunUMAP(st_obj, dims = 1:10, verbose = FALSE)
    st_obj <- Seurat::FindNeighbors(st_obj, dims = 1:10, verbose = FALSE)
    st_obj <- Seurat::FindClusters(st_obj, verbose = FALSE)

    return(st_obj)
}
                                 
n_genes <- 1000
n_cells <- 1000

st_tgt <- random_st_obj(n_genes = n_genes, n_cells = n_cells) # to impute against
st_ref <- random_st_obj(n_genes = n_genes / 2, n_cells = n_cells * 3) # to impute, no need for same size
st_tgt$seurat_clusters <- paste0("tgtcluster", st_tgt$seurat_clusters)
st_ref$seurat_clusters <- paste0("refcluster", st_ref$seurat_clusters)

# Plot the object.
Seurat::DimPlot(st_tgt, label = TRUE, group.by = "seurat_clusters") + 
Seurat::DimPlot(st_ref, label = TRUE, group.by = "seurat_clusters")
                                 
# Get the top marker genes for each.
# Don't use the logfc.
ref_markers <- Seurat::FindAllMarkers(st_ref, logfc.threshold = 0, only.pos = TRUE, verbose = FALSE)
ref_markers <- ref_markers %>% group_by(cluster) %>% slice_max(order_by = avg_log2FC, n = 20)
slice_max(ref_markers, order_by = avg_log2FC, n = 1)
```

```r
# Do the imputation. Since we used random data for each of these, adobo will likely not know
# what our output clusters look like since all of them look the same. This is exepcted.
# With more informative clusters, you'd get 'refclusterX' in the adobo_clusters section.
# THIS TUTORIAL IS JUST TO SHOW HOW TO USE THE FUNCTION.

st_tgt_imputed <- dmSC::adobo_impute(st_tgt,
                                     gene_symbols = ref_markers$gene,
                                     cell_types = ref_markers$cluster,
                                     cluster_slot = "seurat_clusters")

head(st_tgt) # before impute
head(st_tgt_imputed[[]]) # after impute
```

## Single-cell type imputation

A function to easily impute cell types for unlabeled single-cell data.

```r
# Load the reference and target single-cell files.
load("sample.rda")
load("mca3_adult_uterus.rda")

# Give the files sensible names.
tgt_st <- sample
ref_st <- mca3_adult_uterus

# Plot each of the Seurat objects.
saved <- options(repr.plot.height = 8, repr.plot.width = 16)
Seurat::DimPlot(tgt_st, group.by = "seurat_clusters") + Seurat::DimPlot(ref_st, group.by = "cell_lineage")
options(saved)
```

```r
# Perform the single-cell imputation.
tgt_st <- dmSC::impute_st(tgt_st = tgt_st,
                          ref_st = ref_st,
                          out_nm = "singler_clusters_lineages",
                          tgt_clusters_nm = "seurat_clusters",
                          ref_clusters_nm = "cell_lineage")
```

```r
# Show how the imputed cell types have been created.
saved <- options(repr.plot.height = 8, repr.plot.width = 16)
Seurat::DimPlot(tgt_st, group.by = "seurat_clusters") + Seurat::DimPlot(tgt_st, group.by = "singler_clusters_lineages")
options(saved)
```

## Volcano plot generation

A convenience function to get a nice volcano plot.

```r
# Make a sample volcano plot data.frame.
# A rough approximation but it works.
sample_volcano_plot_df <- data.frame(log2FoldChange = runif(n = length(gene_names), min = -5, max = 5),
                                     padj = runif(n = length(gene_names), min = 0, max = 1))
sample_volcano_plot_df$log2FoldChange <- sample_volcano_plot_df$log2FoldChange * (1 - sample_volcano_plot_df$padj)
rownames(sample_volcano_plot_df) <- gene_names
head(sample_volcano_plot_df)
```

```r
# Plot the corresponding volcano plot.
dmSC::volcano_plot(df = sample_volcano_plot_df,
                   logfc_column = "log2FoldChange",
                   logfc_cutoff = log2(1.5),
                   padj_column = "padj",
                   padj_cutoff = 0.05)
```
